```python
#importing the pandas library
import pandas as pd
```

```python
#loading csv file using read_csv and previewing geo description data with head(), specifying 10 records
df1 = pd.read_csv('cells_geo.csv', delimiter=';')
df1.head(10)
```

```python
#dropping unnamed column using drop(). The unnamed column is a 
#duplicate of the indexes so it is not needed
df1 = df1.drop('Unnamed: 0', axis=1)
```

```python
#creating 2 variables with lists of the column names to specify string columns to strip of any
#leading or trailing whitespaces and " character
columns = ['DECOUPZONE','ZONENAME']
columns2 = ['VILLES', 'STATUS', 'LOCALISATION',	'DECOUPZONE',	'ZONENAME',	'REGION',	'AREA',	'CELL_ID', 'SITE_CODE']

df1[columns] = df1[columns].apply(lambda val: val.str.strip('"'))
df1[columns2] = df1[columns2].apply(lambda val: val.str.strip())
df1
```

```python
#checking for duplicates in the geo description data
df1[df1.duplicated()]
```

```python
#dropping the duplicated rows from the geo descripiton data
df1 = df1.drop_duplicates()
```

```python
#checking if there are records where the status is null. These records have
#ids that are not in service therefore not needed for the project
df1[df1['STATUS'].isna()]
```

```python
#deleting the records that are not in service
df1 = df1.dropna(subset=['STATUS'])
```

```python
#loading Telecom dataset1 (CDR 20120507) and previewing the data
df2 = pd.read_csv('Telcom_dataset.csv')
df2.head(10)
```

```python
#renaming the produtc column to its correct spelling - product
#the columns argument in rename() takes the changes in form a dictionary
df2 = df2.rename(columns = {'PRODUTC': 'PRODUCT'})
df2
```

```python
#loading Telecom dataset2
#renaming the column names, eliminating character variations so that they match Telcom dataset 1 to enable 
#appropriatemerging on axis=1
df3 = pd.read_csv('Telcom_dataset2.csv')
df3
df3 = df3.rename(columns = {'DW_A_NUMBER': 'DW_A_NUMBER_INT', 'DW_B_NUMBER' : 'DW_B_NUMBER_INT', 'DATE_TIME': 'DATETIME'})
df3
```

```python
#loading and previewing Telcom dataset 3
df4 = pd.read_csv('Telcom_dataset3.csv')
df4
```

```python
#renaming the column names, eliminating character variations so that they match Telcom dataset 1 to enable 
#appropriatemerging on axis=1
df4 = df4.rename(columns = {'SIET_ID' : 'SITE_ID', 'CELLID' : 'CELL_ID', 'DATE_TIME': 'DATETIME'})
df4
```

```python
#merging the 3 telecom datasets on axis 0 using the concat() function
#ignore index is set to true so that the value of the index can increase in
#a chronological manner along the rows
combined_df = pd.concat((df2,df3,df4), axis=0, ignore_index=True)
combined_df
```

```python
#dropping the columns which are not needed according to the CDR description
#and previewing the change
combined_df = combined_df.drop(['CELL_ON_SITE', 'COUNTRY_A', 'COUNTRY_B'], axis=1)
combined_df.head()
```

```python
#applying the strip() function to remove any leading or trailing whitespaces that may be
#there in the string columns
cols = ['PRODUCT', 'DATETIME','DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'CELL_ID', 'SITE_ID']

combined_df[cols] = combined_df[cols].apply(lambda val: val.str.strip())

combined_df
```

```python
#checking for duplicates in the combined telecoms dataframe
combined_df[combined_df.duplicated()]
```

```python
#dropping duplicate rows from the combined telecoms dataframe
combined_df = combined_df.drop_duplicates()
combined_df
```

```python
cell_ids_days = list(combined_df['CELL_ID'].values)
print(len(cell_ids_days))
site_ids_days = list(combined_df['SITE_ID'].values)
print(len(site_ids_days))
```

```python
#on exploring the geo description data, duplicate site codes point to the same
#city. Since we will be merging the geo description data to the combined telecoms data
#based on the site code, we can create a new dataframe to do this merge that will consist of 
#one record of a particular site and the city it points to.
#dropping rows with duplicate site codes does this
unique_df = df1.drop_duplicates(subset='SITE_CODE')
```
