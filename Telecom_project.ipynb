```python
#importing the pandas library
import pandas as pd
```

```python
#loading csv file using read_csv and previewing geo description data with head(), specifying 10 records
df1 = pd.read_csv('cells_geo.csv', delimiter=';')
df1.head(10)
```

```python
#dropping unnamed column using drop(). The unnamed column is a 
#duplicate of the indexes so it is not needed
df1 = df1.drop('Unnamed: 0', axis=1)
```

```python
#creating 2 variables with lists of the column names to specify string columns to strip of any
#leading or trailing whitespaces and " character
columns = ['DECOUPZONE','ZONENAME']
columns2 = ['VILLES', 'STATUS', 'LOCALISATION',	'DECOUPZONE',	'ZONENAME',	'REGION',	'AREA',	'CELL_ID', 'SITE_CODE']

df1[columns] = df1[columns].apply(lambda val: val.str.strip('"'))
df1[columns2] = df1[columns2].apply(lambda val: val.str.strip())
df1
```

```python
#checking for duplicates in the geo description data
df1[df1.duplicated()]
```

```python
#dropping the duplicated rows from the geo descripiton data
df1 = df1.drop_duplicates()
```

```python
#checking if there are records where the status is null. These records have
#ids that are not in service therefore not needed for the project
df1[df1['STATUS'].isna()]
```

```python
#deleting the records that are not in service
df1 = df1.dropna(subset=['STATUS'])
```

```python
#loading Telecom dataset1 (CDR 20120507) and previewing the data
df2 = pd.read_csv('Telcom_dataset.csv')
df2.head(10)
```

```python
#renaming the produtc column to its correct spelling - product
#the columns argument in rename() takes the changes in form a dictionary
df2 = df2.rename(columns = {'PRODUTC': 'PRODUCT'})
df2
```

```python
#loading Telecom dataset2
#renaming the column names, eliminating character variations so that they match Telcom dataset 1 to enable 
#appropriatemerging on axis=1
df3 = pd.read_csv('Telcom_dataset2.csv')
df3
df3 = df3.rename(columns = {'DW_A_NUMBER': 'DW_A_NUMBER_INT', 'DW_B_NUMBER' : 'DW_B_NUMBER_INT', 'DATE_TIME': 'DATETIME'})
df3
```

```python
#loading and previewing Telcom dataset 3
df4 = pd.read_csv('Telcom_dataset3.csv')
df4
```

```python
#renaming the column names, eliminating character variations so that they match Telcom dataset 1 to enable 
#appropriatemerging on axis=1
df4 = df4.rename(columns = {'SIET_ID' : 'SITE_ID', 'CELLID' : 'CELL_ID', 'DATE_TIME': 'DATETIME'})
df4
```

```python
#merging the 3 telecom datasets on axis 0 using the concat() function
#ignore index is set to true so that the value of the index can increase in
#a chronological manner along the rows
combined_df = pd.concat((df2,df3,df4), axis=0, ignore_index=True)
combined_df
```

```python
#dropping the columns which are not needed according to the CDR description
#and previewing the change
combined_df = combined_df.drop(['CELL_ON_SITE', 'COUNTRY_A', 'COUNTRY_B'], axis=1)
combined_df.head()
```

```python
#applying the strip() function to remove any leading or trailing whitespaces that may be
#there in the string columns
cols = ['PRODUCT', 'DATETIME','DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'CELL_ID', 'SITE_ID']

combined_df[cols] = combined_df[cols].apply(lambda val: val.str.strip())

combined_df
```

```python
#checking for duplicates in the combined telecoms dataframe
combined_df[combined_df.duplicated()]
```

```python
#dropping duplicate rows from the combined telecoms dataframe
combined_df = combined_df.drop_duplicates()
combined_df
```

```python
cell_ids_days = list(combined_df['CELL_ID'].values)
print(len(cell_ids_days))
site_ids_days = list(combined_df['SITE_ID'].values)
print(len(site_ids_days))
```

```python
#on exploring the geo description data, duplicate site codes point to the same
#city. Since we will be merging the geo description data to the combined telecoms data
#based on the site code, we can create a new dataframe to do this merge that will consist of 
#one record of a particular site and the city it points to.
#dropping rows with duplicate site codes does this
unique_df = df1.drop_duplicates(subset='SITE_CODE')
```

```python
#verifying that the site code is not repeated in our new dataframe
unique_df[unique_df.duplicated('SITE_CODE')]
```

```python
#because the aspects from the geo description data most pertinent to merging to the telecoms data are the site code
#and cities, a new data frame is created with only the cities and corresponding site codes
#previewing the first 5 rows
dfud = pd.DataFrame((unique_df[['VILLES', 'SITE_CODE']]))
dfud.head()
```

```python
#determining how many records in the telecoms data have site codes whose city
#appears in the geo description data
site_cell_id = unique_df[['VILLES', 'SITE_CODE']].values
city_count = []

#this for loop iterates through each site id in the telecoms data, and through a
#nested loop through the geo description data, checks if that site id in the telecoms data 
#has a matching site code in the geo description data. if so, the corresponding city name
#for that site id is appended to city count list, and the length of that list is determined
#after the loop

for id in site_ids_days:
  for city_cell in site_cell_id:
    if city_cell[1] == id:
      city_count.append(city_cell[0])
    
print(city_count)
print(len(city_count)) #10239 records from the combined telecoms data frame have a corresponding city in the geo data dataset
```

```python
#creating a new version of the combined telecoms dataframe in which the column site id is renamed to
#site code to enable merging with the geo description data.
#the new dataframe is merged with the dataframe containing cities and their corresponding site codes which was
#created earlier from the geo description dataframe.
#it is a left join, so all records from the combined telecoms are kept at this point, including those which didnt have
#corresponding cities
new_combined = combined_df.rename(columns={'SITE_ID':'SITE_CODE'})
new = new_combined.merge(dfud, how='left', on='SITE_CODE')
new
```

```python
#updating the combined data frame so that it only has telecom records whose site codes have a corresponding city(villes is not null)
#the looping code earlier showed this value to be 10239, which is the number of rows in this new dataframe
new_df = new[~new['VILLES'].isnull()]
new_df
```

```python
#ANALYSIS
#Most used city for the three days


#grouping by cities in the new dataframe to get the count of the different cities. 
#creating a for loop to display the city from this grouping with the highest count in the combined cdr record,
#therefore the most used city over the three days

max_used = new_df.groupby('VILLES')['VILLES'].count()
ind = max_used.index.tolist()
for index in ind:
  if max_used[index] == max_used.max():
    print(f'{index} had the most usage over the 3 days (appeared in cdr { max_used[index]} times).')
```

```python
#Which cities were the most used during business and home hours?

#Home

#creating a function that will slice the date time string to only show the time
def time_section(date_time):
  time = date_time[11:21]
  return time

#specifying new dataframe to use - a copy of the merged geo data and cdr records
new3 = new_df.copy()

#updating the third column in the new dataframe to have the datetime string show only the time
new3['DATETIME'] = new_df['DATETIME'].map(lambda datetime: time_section(datetime), na_action='ignore')
new3
# #creating a new dataframe again to only include times before 9am and after 5 pm
new4 = new3[(new3['DATETIME'] < '09:00:00.0') | (new3['DATETIME'] > '17:00:00.0')]


#grouping by city and showing the count of each city that satisfies the conditions above.
#Getting the maximum count
maxim = new4.groupby('VILLES')['VILLES'].count()
index2 = maxim.index.tolist()
for index in ind:
  if maxim[index] == maxim.max():
    print(f'{index} had the most usage during home hours ({ maxim[index]} times in cdr between 5pm and 9am).')
```

```python
#least used cities for the three days. Repeating procedure in cell above but using min()
used = new_df.groupby('VILLES')['VILLES'].count()
ind = used.index.tolist()
least_used = []
for index in ind:
  if used[index] == used.min():
    least_used.append(index)
print(least_used)
```

```python
#city bringing in most profit over the three days

#group by city and sum up the value column, then display the maximum record
profit = new_df.groupby('VILLES')['VALUE'].sum()
profit
ind5 = profit.index.tolist()
# least_used = []
for index in ind:
  if profit[index] == profit.max():
    print(index, profit[index])
```

```python
#most valuable product over the 3 days where site ids have corresponding city data available

#group by product, sum up the value in terms of billing price
# Display maximum price and corresponding product
products = new_df.groupby(['PRODUCT'])['VALUE'].sum()
products
ind6 = products.index.tolist()

for index in ind6:
   if products[index] == products.max():
     print(f'{index}, with billing price of {products[index]}')
```
